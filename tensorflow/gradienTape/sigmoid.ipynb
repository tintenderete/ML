```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    # Creamos constante y variable:
    x = tf.Variable(x)

    with tf.GradientTape() as tape:
        y = tf.sigmoid(x)

    dy_dx = tape.gradient(y, x)
    

    return y, dy_dx
```

```python
#x = np.arange(-1.0, 1.0, 0.1) # No son suficientes valores
x = np.arange(-10., 10., 0.01)
```

```python
y , dy_dx = f(x)
```

```python
plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(1, 2, 2)
plt.plot(x, dy_dx)
plt.xlabel("x")
plt.ylabel("dy_dx")
plt.show()
```
