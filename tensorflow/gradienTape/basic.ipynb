```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python

def f(a,x):
    # Creamos constante y variable:
    x = tf.Variable(x)

    with tf.GradientTape() as tape:
        y = a*x**2

    dy_dx = tape.gradient(y, x)
    

    return y, dy_dx
```

```python
y , dy_dx = f(1.0,4.0)
print("y =", y)
print("y =", y.numpy())

print("dy_dx =", dy_dx)
print("dy_dx =", dy_dx.numpy())
```

```python
x = np.arange(-1.0, 1.0, 0.1)
print(x)
print(x.shape)
```

```python
y , dy_dx = f(1.0,x)
```

```python
print("y =", y)
#print("y =", y.numpy())
```

```python
print("dy_dx =", dy_dx)
#print("dy_dx =", dy_dx.numpy())
```

```python
plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(1, 2, 2)
plt.plot(x, dy_dx)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
