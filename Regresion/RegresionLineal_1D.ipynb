```python
import numpy as np
import matplotlib.pyplot as plt
```

Creamos una serie de datos para la funcion y = w*x + b

```python
# Parameters:
a = 2.0
b = 1.0
xmin = 0.0
xmax = 10.0
noise = 2.0
n = 100

# Randomly generated problem data:
x = xmin + np.random.rand(n)*(xmax - xmin)
t = a*x + b + np.random.randn(n)*noise # y = w*x+b * RUIDO
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, t, 'o') # Puntos azules
plt.plot([xmin, xmax], [a*xmin + b, a*xmax + b], 'r-') # Linea Roja
plt.grid(True)
plt.xlabel("x")
plt.ylabel("t")
plt.show()
```

Descenso por gradiente:

```python
w = np.random.randn()
b = np.random.randn()

num_iters = 16
eta = 0.0001

plt.figure(figsize=(12,12))

for i in range(num_iters):
    y = w*x + b # Funcion lineal a optimizar
    
    plt.subplot(4, 4, i+1)
    plt.plot(x, t, 'o') # Puntos azules, Datos
    plt.plot([xmin, xmax], [w*xmin + b, w*xmax + b], 'r-') # Linea Roja, Modelo
    plt.xticks([])
    plt.yticks([])
    plt.title("epoch = %d" % i)
    
    # Calculo error - Descenso
    y_minus_t = y - t
    dw = np.sum(y_minus_t*x)
    db = np.sum(y_minus_t)
    w -= eta*dw
    b -= eta*db

plt.show()
```
