```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Crear datos de ejemplo con relación lineal
x_data = np.random.rand(100).astype(np.float32)
y_data = x_data * 2 + 0
```

```python
# Crear variables para los pesos y el bias del modelo
W = tf.Variable(tf.random.normal([1]), name="weight")
b = tf.Variable(tf.zeros([1]), name="bias")
```

```python
# Crear el modelo lineal
def linear_regression(x):
    return W * x + b

# Definir la función de costo (MSE)
def mean_square_error(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred - y_true))

# Definir el optimizador
optimizer = tf.optimizers.SGD(learning_rate=0.01)
```

```python
# Entrenar el modelo
for i in range(2000):
    # Calcular los gradientes
    with tf.GradientTape() as tape:
        y_pred = linear_regression(x_data)
        loss = mean_square_error(y_pred, y_data)
    gradients = tape.gradient(loss, [W, b])
    
    # Actualizar los pesos y el bias
    optimizer.apply_gradients(zip(gradients, [W, b]))
```

```python
# Graficar los datos y la línea de regresión
plt.scatter(x_data, y_data, color='blue')
plt.plot(x_data, linear_regression(x_data), color='red')
plt.show()
```

```python
# Crear una lista de enteros 
lista_enteros = tf.range(1., 100.)
```

```python
plt.plot(lista_enteros, linear_regression(lista_enteros), color='red')
plt.show()
```

```python
# Calcular derivadas

dy_dw_list = []

for i in range(lista_enteros.shape[0]):

  x = tf.Variable(lista_enteros[i])

  with tf.GradientTape() as tape:
    y = linear_regression(x)

  dy_dw = tape.gradient(y,  [W, b])

  dy_dw_list.append(dy_dw[0].numpy()[0])
```

```python
# Graficar derivada
plt.plot(lista_enteros, dy_dw_list, color='brown')
plt.show()
```
